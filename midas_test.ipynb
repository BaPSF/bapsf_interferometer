```python
import h5py
import time
from datetime import datetime

import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
ifn = "/data/BAPSF_Data/Chen/July2024/interferometer/interferometer_data_2024-07-09.hdf5"
f = h5py.File(ifn, "r")
```

```python
phase_20 = f['phase_p20']
phase_29 = f['phase_p29']
tarr_data = f['time_array']

tstamp_ls = list(tarr_data.keys())
print(len(tstamp_ls))
```

```python
fig1, ax1 = plt.subplots()
fig2, ax2 = plt.subplots()
ax1.set_title("port20")
ax1.set_xlabel("time")
ax1.set_ylabel('phase')
ax2.set_title("port29")
ax2.set_xlabel("time")
ax2.set_ylabel('phase')

ts = tstamp_ls[50000]
tarr = tarr_data[ts][:]
ax1.plot(tarr, phase_20[ts][:], label=time.ctime(float(ts)) )
ax2.plot(tarr, phase_29[ts][:], label=time.ctime(float(ts)) )

for ts in tstamp_ls[62000::500]:

    tarr = tarr_data[ts][:]

    ax1.plot(tarr, phase_20[ts][:], label=time.ctime(float(ts)) )
    ax2.plot(tarr, phase_29[ts][:], label=time.ctime(float(ts)) )
    
ax1.legend()
ax2.legend()
plt.show()
```
